```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Dataset (Student Marks Data)
data = {
    'Department': ['CSE', 'ECE', 'MECH', 'CSE', 'ECE', 'MECH', 'CSE', 'ECE', 'MECH', 'CSE'],
    'Marks': [85, 78, 65, 90, 88, 70, 95, 72, 60, 80],
    'Gender': ['Female', 'Male', 'Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female']
}

df = pd.DataFrame(data)

# -----------------------------
# 1️⃣ Bar Chart - Average Marks by Department
# -----------------------------
avg_marks = df.groupby('Department')['Marks'].mean()

plt.figure()
avg_marks.plot(kind='bar')
plt.title("Average Marks by Department")
plt.xlabel("Department")
plt.ylabel("Average Marks")
plt.show()

# -----------------------------
# 2️⃣ Pie Chart - Gender Distribution
# -----------------------------
gender_count = df['Gender'].value_counts()

plt.figure()
gender_count.plot(kind='pie', autopct='%1.1f%%')
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()

# -----------------------------
# 3️⃣ Histogram - Marks Distribution
# -----------------------------
plt.figure()
plt.hist(df['Marks'], bins=5)
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```
