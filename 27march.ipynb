Objective

To identify word pairs and explain their semantic similarity (how closely their meanings are related).

Word Pairs with Explanation
1. Happy – Joyful
Type: Synonyms
Explanation: Both words express the same emotion of happiness and positivity.
 High semantic similarity
2. Car – Vehicle
Type: Category relationship
Explanation: A car is a type of vehicle, so they are related in meaning.
Moderate to high similarity
3. Teacher – Student
Type: Related (contextual relationship)
Explanation: They are not similar in meaning but are connected in an educational context.
Moderate similarity
4. Hot – Cold
Type: Antonyms
Explanation: These words have opposite meanings but belong to the same concept (temperature).
Low similarity (but strongly related conceptually)
5. Dog – Puppy
Type: Same concept (age difference)
Explanation: A puppy is a young dog, so both refer to the same animal category.
 High similarity
Explanation of Semantic Similarity

Semantic similarity refers to:
How closely two words are related in meaning

It can be:

Synonyms → same meaning (happy–joyful)
Antonyms → opposite meaning (hot–cold)
Hierarchical → one is a type of another (car–vehicle)
Contextual → related by usage (teacher–student)
