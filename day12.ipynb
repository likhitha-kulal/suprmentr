```python
# Taking user input
name = input("Enter your name: ")
age = int(input("Enter your age: "))
hobby = input("Enter your hobby: ")

# Categorizing age using conditionals
if age < 13:
    category = "Child"
elif age < 20:
    category = "Teenager"
elif age < 60:
    category = "Adult"
else:
    category = "Senior"

# Displaying personalized message
print("\nHello", name + "!")
print("You are classified as:", category)
print("It's great that you enjoy", hobby + "!")
```
